## Agregar datos por usuario

[Ahora que los datos están limpios, agrega los datos por usuario y por periodo para que solo haya un registro por usuario y por periodo. Esto facilitará mucho el análisis posterior.]

```python
calls_per_user_month = (
    calls.groupby(["user_id", "year_month"])["id"]
    .count()
    .reset_index(name="num_calls")
)
print(calls_per_user_month)# Calcula el número de llamadas hechas por cada usuario al mes. Guarda el resultado.

```

```python
minutes_per_user_month = (
    calls.groupby(["user_id", "year_month"])["duration"]
    .sum()
    .reset_index(name="total_minutes")
)
minutes_per_user_month["total_minutes"] = np.ceil(minutes_per_user_month["total_minutes"]).astype("int")
print(minutes_per_user_month)# Calcula la cantidad de minutos usados por cada usuario al mes. Guarda el resultado.

```

```python
sms_per_user_month = (
    messages.groupby(["user_id", "year_month"])["id"]
    .count()
    .reset_index(name="num_messages")
)
print(sms_per_user_month)   # Calcula el número de mensajes enviados por cada usuario al mes. Guarda el resultado.
```

```python
internet_per_user_month = (
    internet.groupby(["user_id", "year_month"])["mb_used"]
    .sum()
    .reset_index(name="total_mb_used")
)
internet_per_user_month["total_mb_used"] = np.ceil(internet_per_user_month["total_mb_used"]).astype("int")
internet_per_user_month["gb_used"] = np.ceil(internet_per_user_month["total_mb_used"]/1024).astype("int")
print(internet_per_user_month)
# Calcula el volumen del tráfico de Internet usado por cada usuario al mes. Guarda el resultado.

```

[Junta los datos agregados en un DataFrame para que haya un registro que represente lo que consumió un usuario único en un mes determinado.]

```python
user_monthly_data = (
    calls_per_user_month
    .merge(minutes_per_user_month, on=["user_id", "year_month"], how="outer")
    .merge(sms_per_user_month, on=["user_id", "year_month"], how="outer")
    .merge(internet_per_user_month, on=["user_id", "year_month"], how="outer")
)
print(user_monthly_data.head())
# Fusiona los datos de llamadas, minutos, mensajes e Internet con base en user_id y month

```

<div class="alert alert-block alert-danger">
<b>Comentario del revisor (1ra Iteración)</b> <a class=“tocSkip”></a>

Hiciste una correcta agrupación de los datos combinando `groupby`  y `merge`, pero recuerda que, según el enunciado del proyecto, los valores de duración de llamadas y consumo de internet deben redondearse hacia arriba, ya que el cobro se hace por minuto y GB completo. Para esto debes tener en cuenta lo siguiente:

* **Llamadas:** Primero realiza el redondeo hacia arriba de los minutos para cada llamada y luego sumas los minutos totales por usuario para cada mes.

* **Internet:** Primero sumas el consumo total en MB para cada usuario por mes, luego conviertes ese valor a GB y finalmente realizas el redondeo hacia arriba.

Los valores finales que obtienes para cada usuario, los comparas con los minutos y GB incluidos en cada plan, con el fin de establecer si hubo un exceso en el consumo y hacer el cobro respectivo.

</div>

```python
user_monthly_with_plan = user_monthly_data.merge(
    users[["user_id", "plan"]], 
    on="user_id", 
    how="left"
)
user_monthly_with_plan = user_monthly_with_plan.merge(
    plans,
    left_on="plan",   
    right_on="plan_name", 
    how="left"
)

user_monthly_with_plan = user_monthly_with_plan.merge(
    plans[["plan_name", "usd_per_gb"]],
    on="plan_name",
    how="left"
)

user_monthly_with_plan = user_monthly_with_plan.merge(
    users[['user_id', 'city']],
    on='user_id', how='left'
)

user_monthly_with_plan = user_monthly_with_plan.fillna(value=0)

print(user_monthly_with_plan)


# Añade la información de la tarifa




```

[Calcula los ingresos mensuales por usuario (resta el límite del paquete gratuito del número total de llamadas, mensajes de texto y datos; multiplica el resultado por el valor del plan de llamadas; añade la tarifa mensual en función del plan de llamadas). Nota: Dadas las condiciones del plan, ¡esto podría no ser tan trivial como un par de líneas! Así que no pasa nada si dedicas algo de tiempo a ello.]

```python
import numpy as np
user_monthly_with_plan["extra_minutes"] = (user_monthly_with_plan["total_minutes"] - user_monthly_with_plan["minutes_included"]).clip(lower=0)
user_monthly_with_plan["extra_messages"] = (user_monthly_with_plan["num_messages"] - user_monthly_with_plan["messages_included"]).clip(lower=0)
user_monthly_with_plan["extra_mb"] = (user_monthly_with_plan["total_mb_used"] - user_monthly_with_plan["mb_per_month_included"]).clip(lower=0)
user_monthly_with_plan["extra_gb"] = np.ceil(user_monthly_with_plan["extra_mb"]/1024).astype("int")
user_monthly_with_plan["extra_minutes_price"] = (user_monthly_with_plan["extra_minutes"] * user_monthly_with_plan["usd_per_minute"])
user_monthly_with_plan["extra_messages_price"] = (user_monthly_with_plan["extra_messages"] * user_monthly_with_plan["usd_per_message"])
user_monthly_with_plan["extra_gb_price"] = user_monthly_with_plan["extra_gb"]*user_monthly_with_plan["usd_per_gb_y"]

user_monthly_with_plan["total_monthly_cost"] = (
    user_monthly_with_plan["usd_monthly_pay"]
    + user_monthly_with_plan["extra_minutes_price"]
    + user_monthly_with_plan["extra_messages_price"]
    + user_monthly_with_plan["extra_gb_price"]
)
print(user_monthly_with_plan)
```

## Prueba las hipótesis estadísticas

[Prueba la hipótesis de que son diferentes los ingresos promedio procedentes de los usuarios de los planes de llamada Ultimate y Surf.]

[Elabora las hipótesis nula y alternativa, escoge la prueba estadística, determina el valor alfa.]

```python
from scipy import stats as st# Prueba las hipótesis
surf_income = user_monthly_with_plan[user_monthly_with_plan["plan"] == "surf"]["total_monthly_cost"]
ultimate_income = user_monthly_with_plan[user_monthly_with_plan["plan"] == "ultimate"]["total_monthly_cost"]
results = st.ttest_ind(surf_income, ultimate_income, equal_var=False)

alpha = 0.05

print("Estadístico t:", results.statistic)
print("p-valor:", results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

```python
nynj_income = user_monthly_with_plan[user_monthly_with_plan["city"].isin(["New York-Newark-Jersey City, NY-NJ-PA MSA"])]["total_monthly_cost"]

other_income = user_monthly_with_plan[~user_monthly_with_plan["city"].isin(["New York-Newark-Jersey City, NY-NJ-PA MSA"])]["total_monthly_cost"]


results = st.ttest_ind(nynj_income, other_income, equal_var=False)

print("Estadístico t:", results.statistic)
print("p-valor:", results.pvalue)

alpha = 0.05
if results.pvalue < alpha:
    print("Rechazamos H0 → Hay diferencia significativa entre los ingresos promedio.")
else:
    print("No rechazamos H0 → No hay evidencia suficiente de diferencia en los ingresos.")# Prueba las hipótesis

```
